```python
inp_layer = lay.Input(shape=(21, )) #Input layer

int_layer = lay.Dense(30, activation='linear')(inp_layer) #Intermediate layer

out_layer = lay.Dense(1, activation='sigmoid')(int_layer) #Output layer

NN = Model(inputs=inp_layer, outputs=out_layer)
```

```python
NN.summary()
```

```python
NN.compile(optimizer=ker.optimizers.Adam(learning_rate=0.01), loss='binary_crossentropy', metrics='accuracy')
```
